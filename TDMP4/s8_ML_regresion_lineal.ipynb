# Regresión

Las regresiones es una de las herramientas principales de la estadistica inferencial. 

El objetivo del análisis de regresión es describir la relación entre un conjunto de variables, llamadas __variables dependientes__, y otro conjunto de variables, llamadas __variables independientes o explicativas__. Se llama regresión lineal simple cuando solo hay una variable independiente y regresión lineal múltiple cuando hay más de una.

>El análisis de regresión ayuda a entender cómo el valor de la variable dependiente cambia cuando cualquiera de las variables independientes es cambiada.

El producto final del análisis de regresión es la estimación de una función de las variables independientes llamada la __función de regresión__. La idea es que en base a esta función de regresión podamos hacer estimaciones sobre eventos futuros.


## Regresión lineal 

La __regresión lineal__ es una de las técnicas más simples y mayormente utilizadas en los análisis de regresiones. 

* Una variable dependiente $Y$
* Variable(s) independiente(s) $X_1, \dots , X_K$.




<img src="./images/rl.PNG" alt="Supervisado" style="width: 400px;"/> 



__Objetivo__: Se busca una función de regresión: $Y=β_0+β_1∗X_1 + β_2∗X_2 + ... + β_k∗X_k $







* Cuando k=1, la función de regresión  es  una línea recta, 
* cuando k=2, la  función de regresión  es  un plano y, por último, 
* cuando k>2,  la función de regresión  es denominada genéricamente hiperplano.





La idea básica de la regresión consiste en estimar los parámetros 
$$β_0,  β_1, ... , β_k $$  a partir de una muestra dada.


Por norma general, los estudios de correlación lineal preceden a la generación de modelos de regresión lineal. Primero se analiza si ambas variables están correlacionadas y, en caso de estarlo, se procede a generar el modelo de regresión.

### Cálculo de los parámetros  $β_i$


Uno de los métodos más populares para realizar regresiones lineales es el de __mínimos cuadrados ordinarios__ (OLS, por sus siglas en inglés).
Este método es el estimador más simple y común en la que los $β_i$s se eligen para minimizar el cuadrado de la distancia entre los valores estimados y los valores reales.

<img src="./images/error_mco.PNG" alt="Supervisado" style="width: 300px;"/>



### Condiciones del modelo de regresión lineal

Para que un modelo de regresión lineal por mínimos cuadrados, y las conclusiones derivadas de él, sean completamente válidas, se deben verificar que se cumplen algunas condiciones. En la práctica, rara vez se cumplen, o se puede demostrar que se cumplen todas, sin embargo esto no significa que el modelo no sea útil. Lo importante es ser consciente de ellas y del impacto que esto tiene en las conclusiones que se extraen del modelo. 

1. __Ausencia de valores atípicos__. Es importante identificar observaciones que sean atípicas o que puedan estar influenciando al modelo. La forma más fácil de detectarlas es a través de los residuos (ver más detalles adelante). 

<img src="./images/reg_err.PNG" alt="Supervisado" style="width: 600px;"/>

2. Los __residuos__ deben seguir una __distribución normal__. Los residuos son las distancias entre los puntos reales y la recta de regresión. 

3. Varianza constante de la variable respuesta (__homocedasticidad__). La varianza de $Y$  debe ser constante en todo el rango de variables independientes. Para comprobarlo suelen representarse los residuos del modelo frente a cada predictor.

4. Precisa __normalización de los datos__.

5. No __multicolinealidad__ entre las variables independientes $X_i$. En caso de multicolinealidad, no se puede identificar de forma precisa el efecto individual que tiene cada $X_i$ sobre la variable $Y$. No es fácil demostrar si ocurre. 

6. Tamaño de la muestra: No se trata de una condición de por sí pero, si no se dispone de suficientes observaciones, las variables $X_i$ que no son realmente influyentes podrían parecerlo. Una recomendación frecuente es que el número de observaciones sea como mínimo entre 10 y 20 veces el número de variables independientes del modelo. 

7. La variable $Y$ tiene una distribución de tipo normal. 


La gran mayoría de condiciones se verifican utilizando los residuos, por lo tanto, se suele generar primero el modelo y posteriormente validar las condiciones. De hecho, el ajuste de un modelo debe verse como un proceso iterativo en el que se ajusta el modelo, se evalúan sus residuos y se mejora. Así hasta llegar a un modelo óptimo.


## Bondad de ajuste del modelo

Las métricas más utilizadas para medir la calidad del modelo son los que se detallan a continuación:


### Error absoluto medio (MAE)

Mide la desviación media de la distancia entre cualquier punto estimado por el modelo respecto de la recta de regresión. Tiene las mismas unidades que la variable $Y$.
<img src="./images/mae.PNG" alt="Supervisado" style="width: 300px;"/>

### Error cuadrático medio (MSE)

Mide la media del cuadrado de las distancias entre cualquier punto estimado por el modelo respecto de la recta de regresión. Tiene las mismas unidades que la variable $Y^2$. Esta métrica da importancia a los outliers. Esto implica que el error crece en presencia de outliers.

<img src="./images/mse.PNG" alt="Supervisado" style="width: 300px;"/>



### Raiz del Error cuadrático medio (RMSE)

Raiz cuadrada de MSE. Tiene las mismas unidades que la variable $Y^2$.

Esta métrica da importancia a los outliers.  Bajo ciertas suposiciones (Sesgo 0, normalidad de los residuos) se puede asegurar con un 95% de probabilidades que el valor real está en el intervalo [-1.96 x RMSE , 1.96 x RMSE]

### Coeficiente  $R^2$

Describe la proporción de varianza de la variable $Y$ explicada por el modelo y relativa a la varianza total. Solo tiene sentido para regresión lineal. Su valor está acotado entre 0 y 1. Cuanto más cercano a 1 mejor será el modelo.

__Construcción de un modelo de regresión lineal__

Construcción de un modelo de regresión lineal para predecir la nota de matemáticas `MAT` en función de la renta per cápita  (`RPC`).

¿Cómo es la distribución de la variable a predecir? `MAT`
Usamos el test de Kolmogorov para estudiar la normalidad:
```
H0: la distribución tiene una distribución normal.
```

````python
resume_MAT = df.select([df.MAT]).describe().toPandas()
resume_MAT
````

````python
from pyspark.ml.stat import KolmogorovSmirnovTest
import pyspark.sql.functions as f
media = float(resume_MAT[resume_MAT.summary == 'mean'].MAT.iloc[0])
std = float(resume_MAT[resume_MAT.summary == 'stddev'].MAT.iloc[0])
ksResult = KolmogorovSmirnovTest.test(df, 'MAT', 'norm', media, std).first()

alpha = 0.05                              # 95% de confianza
if ksResult.pValue < alpha:  # null hypothesis: x comes from a normal distribution
    print(f"The null hypothesis can be rejected ({ksResult.pValue})")
else:
    print(f"The null hypothesis cannot be rejected ({ksResult.pValue})")
````

* Paso 1: Preparar los datos para el modelo. Tenemos que reunir todas las variables independientes (en este caso la variable 'RPC') en un vector. Para ello usamos `Vector Assembler`.

````python
# Importamos la librería 
from pyspark.ml.regression import LinearRegression    # método de regresión lineal
from pyspark.ml.feature import VectorAssembler        # para construir un vector con los datos de las variables independientes


# generamos un vector a partir de los valores de las diferentes variables independientes del modelo (X_i)
# y podremos llamar a la nueva columna de vectores como "Atributos" en outputCol.
# en nuestro caso solo tenemos la variable 'RPC'
assembler = VectorAssembler( inputCols = ['RPC'],
                             outputCol ='Attributes')

output =  assembler.transform(df)
output.show(4)
````

````python
data_vector = output.select("Attributes","MAT")     # estos son los datos para el modelo
data_vector.show(4)
````

* Paso 2: Dividir los datos en conjunto de entrenamiento y test (0.8 y 0.2 en este caso). Para ello usamos el método `randomSplit`. El resultado son 2 dataframes:

````python
train, test = data_vector.randomSplit([0.8, 0.2])
train.count(), test.count()
````

Es importante verificar que la distribución de la variable `Y` es similar en el conjunto de entrenamiento y en el de test. 

````python
test.select(['MAT']).describe().show()
````

````python
train.select(['MAT']).describe().show()
````

* Paso 3: Preparar el método de regresión lineal y entrenarlo para obtener el modelo. El modelo se llamará `modelo_lr`. 

````python
# preparamos el método
regressor = LinearRegression(featuresCol = 'Attributes', labelCol = 'MAT')

# entrenamos el modelo con los datos de entrenamiento
modelo_lr = regressor.fit(train)
````

* Paso 4: Usamos los datos de test para evaluar el modelo y hacer predicciones

````python
# predecir las notas sobre los datos de test
test_pred = modelo_lr.transform(test)   # hace las predicciones sobre el conjunto de test
````

````python
test_pred.show(5)
````

__Evaluación del modelo__

Calcularemos el valor de $R^2$ y el error cuadrático medio RMSE (solo estos). 


````python
from pyspark.ml.evaluation import RegressionEvaluator

# predecir las notas sobre los datos de test (medimos la varianza)
evaluadorRMSE = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="rmse") 
rmse = evaluadorRMSE.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test

evaluadorR2 = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="r2") 
r2 = evaluadorR2.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test
print(f'R^2: {r2}')
print(f'Raíz cuadrada del error cuadráticao medio (RMSE): { rmse}')
````

## Hiperparámetros

Los hiperparámetros configuran el método a emplear  y pueden tener un gran impacto en su precisión. 
Cada técnica tiene su lista de hiperparámetros que conviene conocer (el valor por defecto no siempre es el mejor). 

A continuación discutimos los hiperparámetros que se pueden usar cuando sospechamos de overfitting.


En casos de overfitting, es decir, cuando el modelo se comporta bien sobre los datos de entrenamiento, pero no sobre los datos de test (alta varianza), se pueden aplicar __técnicas de regularización__.

### Ridge regression (Regresión lineal contraída)

Supongamos que nos encontramos en la situación de la imagen siguiente:
<table>
    <tr>
        <td><img src="./images/ridgeR.PNG" alt="Supervisado" style="width: 200px;"/></td>
        <td>
            Puntos rojos: representan el conjunto de datos de entrenamiento.<br>
            Puntos verdes: representan el conjunto de datos de test.<br>
            Recta roja: recta de regresión proporcionada por el modelo.<br>
            <br><br><br><br>
            Sesgo: El error medio sobre el conjunto de entrenamiento es 0.<br>
            Varianza: error medio sobre el conjunto de test es alto. <br>
        </td>
     </tr>
</table>

Nos encontramos ante un caso de overfitting (sobreajuste). El objetivo es crear un modelo de regresión que permita aumentar el sesgo y disminur la varianza. Es decir, reducir los efectos de la variable X sobre la variable Y.

<td><img src="./images/lr_ridge.PNG" alt="Supervisado" style="width: 300px;"/></td>

Para ello, se introduce un hiperparámetro $\lambda \ge 0$ que minimiza el error cuadrático medio. Se trata de minimizar la función de pérdida):
$$
\sum (Y_i - Y_i^{'})^2 + \lambda  \beta_i^2
$$
La idea es reducir los coeficientes para aquellas variables de entrada que no contribuyen mucho a la tarea de predicción. __Ridge Regression__ es una extensión de la regresión lineal que agrega una penalización de regularización a la función de pérdida durante el entrenamiento. Se corresponde con una regularización de tipo L2 (ridge regression).
Si  $\lambda = 0$, nos encontramos ante un ejemplo de regresión linea pura, sin regularización.

Este algoritmo de regresión es útil cuando hay pocos datos de entrenamiento o cuando hay más variables independientes que número de observaciones.

Vamos a ejecutar nuestro ejemplo con distintos valores del hiperparámtro $\lambda$ que permite agregar la penalización 

````python
# preparamos el método de ridge regresion con lambda = 0.7
regressor = LinearRegression(featuresCol = 'Attributes',
                             labelCol = 'MAT', 
                             regParam=1,
                             elasticNetParam=0)                  #ridge regresion

# entrenamos el modelo con los datos de entrenamiento
modelo_lr = regressor.fit(train)
# predecir las notas sobre los datos de test
test_pred = modelo_lr.transform(test)   # hace las predicciones sobre el conjunto de test

#evaluar
evaluadorRMSE = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="rmse") 
rmse = evaluadorRMSE.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test
evaluadorR2 = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="r2") 
r2 = evaluadorR2.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test


print(f'R^2: {r2}')
print(f'Raíz cuadrada del error cuadráticao medio (RMSE): { rmse}')
````

Para ajustar el valor del hiperparámetro $\lambda$, podemos crear y entrenar el modelo con varios valores y seleccionar el que mejor resultado nos proporcione:

````python
import numpy as np
lr2 = []
lmse = []
x = np.linspace(0, 5, 50)
for val_lambda in np.linspace(0, 5, 50):
    # preparamos el método de ridge regresion con lambda = 0.7
    regressor = LinearRegression(featuresCol = 'Attributes', labelCol = 'MAT', 
                                 regParam=val_lambda,
                                 elasticNetParam = 0)

    # entrenamos el modelo con los datos de entrenamiento
    modelo_lr = regressor.fit(train)
    # predecir las notas sobre los datos de test
    test_pred = modelo_lr.transform(test)   # hace las predicciones sobre el conjunto de test
    evaluadorRMSE = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="rmse") 
    rmse = evaluadorRMSE.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test
    evaluadorR2 = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="r2") 
    r2 = evaluadorR2.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test
    #print(f'lambda: {val_lambda}')
    #print(f'   R^2: {pred.r2}')
    #print(f'   Error cuadrático medio (MSE): { pred.meanSquaredError}')
    lr2.append(r2)
    lmse.append(rmse)
    
    
fig, ax = plt.subplots(figsize =(7, 4))
ax.plot(x, lr2, color = 'red')
ax.set_ylabel('R^2', color='red')  # we already handled the x-label with ax1
ax.tick_params(axis='y', labelcolor='red')
ax.set_xlabel('Valor de $\lambda$')
ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('mse', color='blue')  # we already handled the x-label with ax1
ax2.plot(x, lmse, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
````

Parece que los mejores valores se obtinen  con $\lambda = 0$, por lo que en este caso, regularizar no ayuda. (No se trata de un caso de overfitting)

### Lasso regresion

Otra forma de regularización para evitar el overfitting, de  reducir los efectos de la variable X sobre la variable Y es minimizar la siguiente función:


$$
\sum (Y_i - Y_i^{'})^2 + \lambda \mid \beta_i \ \mid
$$

Igual que en Ridge Regression, la idea es reducir los coeficientes para aquellas variables de entrada que no contribuyen mucho a la tarea de predicción. Se corresponde con una regularización de tipo L1 (Lasso regression).

__L1 permite seleccionar variables ya que tiende a asignar coeficiente 0 a las variables no utilizadas.__

Ambas regularizacines se pueden combinar en un método denominado __elasticNet__. Si el hiperparámetro `elasticNetParam` (con valores entre 0 y 1), toma el valor 1, es equivalente a un modelo `Lasso regression`. Por otro lado, si el hiperparámetro `elasticNetParam` se establece en 0, el modelo entrenado se reduce a un modelo de `ridge regresion`.

````python
# preparamos el método de ridge regresion con lambda = 0.7
regressor = LinearRegression(featuresCol = 'Attributes', labelCol = 'MAT', 
                             regParam=0.7, 
                             elasticNetParam = 1)                  #lasso regresion

# entrenamos el modelo con los datos de entrenamiento
modelo_lr = regressor.fit(train)
# predecir las notas sobre los datos de test
test_pred = modelo_lr.transform(test)   # hace las predicciones sobre el conjunto de test
evaluadorRMSE = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="rmse") 
rmse = evaluadorRMSE.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test
evaluadorR2 = RegressionEvaluator(predictionCol="prediction",  labelCol="MAT", metricName="r2") 
r2 = evaluadorR2.evaluate(test_pred)   # evaluación del modelo sobre el conjunto de test

print(f'R^2: { r2}')
print(f'Raiz cuadrada del error cuadrático medio (RMSE): { rmse}')
````

## Analizando el error

### Caso 1: gran varianza y bias pequeño

Ejemplo:

> Training: 1% error (Bias)
>
> val/test: 11 % error (Varianza = 11-1 = 10%) 

* Posible overfitting
* El error pequeño sobre el conjunto de entrenamiento es engañoso; el algoritmo de alguna forma, en lugar de generalizar a partir del conjunto de entrenamiento, se ha aprendido de memoria los datos de entrenamiento

¿Qué hacer si hay overfitting?

1. ¿Incluye el conjunto de entrenamiento al menos algunos valores similares a los del test? Si no es así, deberíamos incluirlos. Tal vez el conjunto de test sea demasiado pequeño o particular.
2. Añadir más datos al conjunto de entrenamiento
3. Reducir columnas (manualmente o PCA)
4. Utilizar parámetros de “regularización” para evitar el sobreajuste (por ejemplo ridge)

### Caso 2:  gran sesgo y varianza pequeña

Ejemplo:

> Training: 15% error (Bias)
>
> val/test: 17 % error (Varianza 17 - 15  = 2%) 

* Posible underfitting: el modelo no ha aprendido con los datos de entrenamiento.
* Parece que el modelo generaliza bien, no añade error, debemos centrarnos en reducir, si es possible, el sesgo en el conjunto de entrenamiento

¿Qué hacer si hay underfitting?

1. Buscar otros modelos más robustos (redes con más capas/neuronas, random forest con más árboles, métodos con más iteraciones)
2. Añadir más datos al training test no ayuda!!
3. Quitar/atenuar parámetros de regularización (ojo, aumenta la varianza)
4. Añadir columnas o nuevas categorías

Observaremos cómo evoluciona el error RMSE sobre el conjunto de train y el conjunto de test.

````python
# 1. generar un vector a partir de los valores de vectorcols, con nombre 'Attributes'
# obtener un nuevo dataframe `data_vector` que solo incluya 'SCI' y el nuevo vector 'Attributes'
assembler = VectorAssembler(
    inputCols= ['MAT'],
    outputCol='Attributes')

output =  assembler.transform(df)

data_vector = output.select("Attributes","SCI")     # estos son los datos para el modelo
data_vector.show(4)
````

````python
ratio_train = []
rmse_train = []   # lista de errores medios sobre los datos de entrenamiento
rmse_test = []    # lista de errores medios sobre los datos de test
n = 30     
# nombres de columna que no sean label (label es el nombre en Spark para la y)
for ratio in [i/n for i in range(1,n)]: 
    print(f'ratio: {ratio}')
    ratio_train.append(ratio)  
    
    ytr = []
    yte = []
    val_r2 = []
    veces = 5
    for i in range(veces):
        # 2. Dividir `data_vector` en conjunto de entrenamiento 
        train_df, test_df = data_vector.randomSplit([ratio, 1-ratio]) 

        # 3. Declarar el método de regresión lineal y a partir de él y de los datos de entrenamiento el modelo,
        # al que se debe llamar modelo_lr

        regressor = LinearRegression(featuresCol = 'Attributes', labelCol = 'SCI')
        # entrenamos el modelo con los datos de entrenamiento
        modelo_lr = regressor.fit(train_df)

        # Resumen de lo bueno que es el modelo sobre el propio conjunto de entrenamiento
        trainingSummary = modelo_lr.summary
        ytr.append(trainingSummary.rootMeanSquaredError)

        # 4. Mostrar los valores r2 y rmse para el test con el fin de evaluar el modelo
        lr_predictions = modelo_lr.transform(test_df)
        metrics=["rmse", "r2"]


        evaluators = [(m,RegressionEvaluator(predictionCol="prediction", \
                     labelCol="SCI",metricName=m)) for m in metrics]

        errors = [ (name, evalu.evaluate(lr_predictions)) for name,evalu in evaluators]
        yte.append(errors[0][1])  
        val_r2.append(errors[1][1])
    rmse_test.append(np.mean(yte))    # media de error del modelo sobre los datos de test (varianza)
    rmse_train.append(np.mean(ytr))   # media de error del modelo sobre los datos de train (bias)
    
    print("Sesgo (train) ", np.mean(ytr),"Varianza (test) ", np.mean(yte), "r2: ", np.mean(val_r2) )
````

Ahora representamos gráficamente la evolución del error.

````python
%matplotlib inline

import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.plot(ratio_train,rmse_train,label="sesgo (bias) - Error sobre datos del entrenamiento")
ax.plot(ratio_train,rmse_test,label="varianza - Error sobre datos de test")
ax.set_ylabel('Error mrse')
ax.set_xlabel('Valor del ratio')
plt.legend()
plt.show()

````

* El error no va a mejorar por mucho que obtengamos más datos, ambas curvas están estables en una asíntota
* Si queremos usar este modelo basta con poner aprox. el 30% de los datos en el train, parece que poner muchos más no aporta demasiado.
* De tener algún problema este es underfitting y no overfitting; el sistema genera un modelo demasiado simple que ya no es capaz de aprender más con los nuevos datos.
* Por lo anterior, parece que probar a cambiar hiperparámetros (regularización, etc.) no ayudará

Sugerencias si el error no es satisfactorio:

* Cambios en el preprocesamiento para quitar ruido (mejores datos). Quizás quitar dimensiones (aquí no es posible)
* Probar otros modelos más complejos (regresiones no lineales, por ejemplo).
